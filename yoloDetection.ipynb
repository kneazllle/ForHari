```python
import torch
import pandas as pd
from matplotlib import pyplot as plt
from PIL import Image

# Load YOLOv5 model (pretrained on COCO dataset)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Load an image
img = '/Users/anandhu/Documents/miniProject/002.jpeg'  # Replace with your image path

# Perform inference
results = model(img)

# Display results
results.show()

# Extract results
df = results.pandas().xyxy[0]  # Pandas DataFrame of results

# Display the coordinates of each detected object
for index, row in df.iterrows():
    x_start = int(row['xmin'])
    y_start = int(row['ymin'])
    x_length = int(row['xmax'] - row['xmin'])
    y_length = int(row['ymax'] - row['ymin'])
    
    print(f"Object {index + 1}:")
    print(f"  - Class: {row['name']}")
    print(f"  - Confidence: {row['confidence']:.2f}")
    print(f"  - x_start: {x_start}")
    print(f"  - y_start: {y_start}")
    print(f"  - x_length: {x_length}")
    print(f"  - y_length: {y_length}")
    print()
```
